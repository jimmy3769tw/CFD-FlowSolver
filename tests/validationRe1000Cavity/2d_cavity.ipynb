```python
import matplotlib.pyplot as plt

# 读取数据文件
filename = '../../build/Information/y_plot_re_1000.dat'

# 数据点
points = [
    [1.0000, 1.00000],
    [0.9766, 0.65928],
    [0.9688, 0.57492],
    [0.9609, 0.51117],
    [0.9531, 0.46604],
    [0.8516, 0.33304],
    [0.7344, 0.18719],
    [0.6172, 0.05702],
    [0.5000, -0.06080],
    [0.4531, -0.10648],
    [0.2813, -0.27805],
    [0.1719, -0.38289],
    [0.1016, -0.29730],
    [0.0703, -0.22220],
    [0.0625, -0.20196],
    [0.0547, -0.18109],
    [0.0000, 0.00000]
]

x = [point[0] for point in points]
y = [point[1] for point in points]

with open(filename, 'r') as file:
    lines = file.readlines()

# 解析数据
variables = lines[0].strip().split('\t')
data_lines = lines[1:]

x2 = []
y2 = []

for line in data_lines:
    line = line.strip().split('\t')
    x2.append(float(line[0]))
    y2.append(float(line[1]))

# 绘制图形
plt.plot(x2, y2, 'b-')
plt.scatter(x, y, color='black')
plt.grid(True)
plt.show()
```
